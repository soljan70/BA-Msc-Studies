```python
#Daty zlecenia, początku i końca

import datetime
random.seed(420)

def date_range(start, end):
    r = (end+datetime.timedelta(days=1)-start).days
    return [start+datetime.timedelta(days=i) for i in range(r)]
start1 = datetime.date(2020,1,1)
end1 = datetime.date(2020, 12,31)
daterange = date_range(start1, end1)
start_zlecenia = []
start_pracy= []
koniec_pracy = []

for i in range(0,31):
    start_zlecenia.append((random.choice(daterange)))
    a = start_zlecenia[i]+datetime.timedelta(days=60)
    start_pracy.append(random.choice(date_range(start_zlecenia[i], a)))
    if zakres_zlecenia[i] == "remont ogólny":
        b = random.choice(date_range(start_pracy[i]+datetime.timedelta(days=21), start_pracy[i]+datetime.timedelta(days=90)))
    if zakres_zlecenia[i] == "remont łazienki":
        b = random.choice(date_range(start_pracy[i]+datetime.timedelta(days=7), start_pracy[i]+datetime.timedelta(days=21)))
    if zakres_zlecenia[i] == "remont kuchni":
        b = random.choice(date_range(start_pracy[i]+datetime.timedelta(days=10), start_pracy[i]+datetime.timedelta(days=30)))
    if zakres_zlecenia[i] == "remont pokoju":
        b = random.choice(date_range(start_pracy[i]+datetime.timedelta(days=4), start_pracy[i]+datetime.timedelta(days=10)))
    if zakres_zlecenia[i] == "malowanie":
        b = random.choice(date_range(start_pracy[i]+datetime.timedelta(days=2), start_pracy[i]+datetime.timedelta(days=10)))
    if zakres_zlecenia[i] == "elewacja":
        b = random.choice(date_range(start_pracy[i]+datetime.timedelta(days=14), start_pracy[i]+datetime.timedelta(days=45)))

    if b > datetime.date(2020, 12, 31):
                      koniec_pracy.append(datetime.date(9999, 12, 31))
    else:
                      koniec_pracy.append(b)
```
