## Summary

- CDT: the most simple model shows very weak accuracy, most of the tag's classes have not been recognised at all. The metric macro avg shows the summary of the model which equals 0.11.

- CRF_1: the first approach of CRF already shows a quite good efficiency of this method in these kind of problems in NLP. The prediction on the train sample is almost perfect, all the tag's classes have f1-score around 0.99, tag = 'time' was classified with the lowest recall. Metric macro avg was here 0.98.
In the valid sample the results of the metrics were much more diverse. f1-score in some classes was the value in the range (0.2-0.5), macro avg equaled here **0.68**.


- CRF_final: after CV and hyperparameters optimization the results were much more better, especially on the valid and test samples. Classes "time" and "bloc" had the lowest recalls which implicated to the f1-score. "time" class is of the most complex structures of tag to predict and is complicated, so that is the reason propably. All the 13 other classes had f1-score metric in the range 0.98-1.0 in both samples. On the valid sample macro avg was **0.9968**. **That method of prediction show best way at dealing with the NER issue due to dedicated NLP solutions based on a set of appropriate features included in the learning stage.**

- BERT_basic: the deep learning approach did not bring the expected results, comparing to CRF. The base BERT model on the valid sample had values of f1-score in the range (0.05-0.88). Best result received classes: "forename", "surname" and "country". Comparing on the avarage results also "date" class looks good. Macro avg equals **0.64**

- BERT_polish: analogous to the results of BERT_basic with a bit higher values. Macro avg equals **0.65**. Definietely helped the model based and trained on the polish language corpus. Another disadvantage of BERT approach is time of learning the networks (3-4h vs 0.4h in CRF) and hardware resources.
